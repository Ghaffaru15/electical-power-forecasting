```python
from pyforest import *
```

```python
data = pd.read_csv('../../data/cleaned-data/FullData-Sunyani.csv')
```

```python
data.isnull().sum()
```

```python
data = data[['Hour', 'Day', 'Month', 'Pressure', 'Relative Humidity', 'Temperature', 'BF12', 'BF7', 'F8B']]
```

```python
data.head()
```

```python
data['Pressure'] = data['Pressure'].fillna(data['Pressure'].mean())
```

```python
data['Relative Humidity'] = data['Relative Humidity'].fillna(data['Relative Humidity'].mean())
```

```python
data['Temperature'] = data['Temperature'].fillna(data['Temperature'].mean())
```

```python
data['BF12'] = data['BF12'].fillna(data['BF12'].mean())
data['BF7'] = data['BF7'].fillna(data['BF7'].mean())
data['F8B'] = data['F8B'].fillna(data['F8B'].mean())
```

```python
data.isnull().sum()
```

```python
data.head(24)
```

```python
data.shape
```

```python
newdf = pd.DataFrame({}, columns =['Day', 'Month', 'Pressure', 'Humidity', 'Temperature', 'BF12', 'BF7', 'F8B'] )
```

```python
newdf
```

```python
# pressure = 0
# humidity = 0
# temperature = 0
# bf12 = 0
# bf7 = 0
# f8b = 0
for i in range(data.shape[0]):

    if data.iloc[i]['Hour'] == 1:
            newdf = pd.concat([newdf, pd.DataFrame({'Day': data.iloc[i]['Day'], 'Month': data.iloc[i]['Month'],
                                               'Pressure': pressure / 24,
                                               'Humidity': humidity / 24,
                                               'Temperature': temperature / 24,
                                               'BF12': bf12 ,
                                               'BF7': bf7 ,
                                               'F8B': f8b }, index=[i])])

            pressure = data.iloc[i]['Pressure']
            humidity = data.iloc[i]['Relative Humidity']
            temperature = data.iloc[i]['Temperature']
            bf12 = data.iloc[i]['BF12']
            bf7 = data.iloc[i]['BF7']
            f8b = data.iloc[i]['F8B']
  
            
    else:
            pressure = pressure + data.iloc[i]['Pressure']
            humidity = humidity +  data.iloc[i]['Relative Humidity']
            temperature = temperature + data.iloc[i]['Temperature']
            bf12 = bf12 + data.iloc[i]['BF12']
            bf7 = bf7 + data.iloc[i]['BF7']
            f8b = f8b + data.iloc[i]['F8B']
           
```

```python
newdf
```

```python
newdf['Power'] = newdf['BF12'] + newdf['BF7'] + newdf['F8B']
```
