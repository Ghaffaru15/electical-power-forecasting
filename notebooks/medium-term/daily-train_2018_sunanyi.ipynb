```python
features = daily_df[['DAYS','MONTHS','TEMPERATURE','HUMIDITY']]
label = daily_df[['HISTORIC LOAD (MW)']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
features_train, features_test, labels_train, labels_test = train_test_split(features, label, test_size=0.1)
```

```python
features_train.shape
```

```python
labels_train.shape
```

```python
labels_train.iloc[0]
```

```python
from sklearn.svm import LinearSVR
```

```python
model = LinearSVR(epsilon=1.5)
```

```python
model.fit(features_train,labels_train)
```

```python
labels_test.head(10)
```

```python
model.predict([features_test.loc[39
                                ]])
```

```python
plt.plot(np.arange(1,10), labels_test['HISTORIC LOAD (MW)'])
plt.plot(np.arange(1,10), model.predict(features_test))
plt.legend(['Actual','Predicted'])
plt.show()
```

```python
plt.figure(figsize=(20,12))
plt.plot(np.arange(1,82), labels_train['HISTORIC LOAD (MW)'])
plt.plot(np.arange(1,82), model.predict(features_train))
plt.legend(['Actual','Predicted'])
plt.show()
```
